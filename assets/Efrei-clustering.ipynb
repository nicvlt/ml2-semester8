```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#from scipy import linalg
from scipy.cluster.hierarchy import fcluster, dendrogram, linkage
from matplotlib.colors import ListedColormap
from sklearn import cluster
import scipy.io as sio
%matplotlib inline
```

```python
from scipy.cluster.hierarchy import dendrogram

def plot_dendrogram(model, **kwargs):
    distance = np.arange(model.children_.shape[0])
    position = np.arange(2, model.children_.shape[0]+2)

    linkage_matrix = np.column_stack([model.children_, distance, position]).astype(float)

    fig, ax = plt.subplots(figsize=(15, 7))

    ax = dendrogram(linkage_matrix, orientation='top', **kwargs)

    plt.tick_params(axis='x', bottom='off', top='off', labelbottom='off')
    plt.tight_layout()
    plt.show()

def plot_clusters(X, Y, title = " ", symbolsize=20):
    plt.figure(figsize=(8,6))
    plt.scatter(X[:,0], X[:, 1], c=Y,  s=symbolsize, cmap=plt.cm.Set1)
    plt.title(title)
    plt.colorbar(ticks=np.unique(Y))
```

```python
df=pd.read_csv('ligue1_17_18.csv', delimiter=';')
df.info()
```

```python
allmeans=df.loc[:,"Points":"fouls.against.sum"].mean()
allstd= df.loc[:,"Points":"fouls.against.sum"].std()
numericaldf=df.loc[:,"Points":"fouls.against.sum"]
```

```python
ligue1_scaled=(numericaldf-allmeans)/allstd
```

```python
ligue1_scaled
```

```python
Y=df.Team
```

```python
from scipy.spatial.distance import pdist
# Compute dissimilarity matrix - in this case Euclidean distance
d = pdist(ligue1_scaled, metric='euclidean')

# Hierarchical clustering with complete linkage
treeComp = linkage(d, method='complete')

# Plot the dendrogram
plt.figure(figsize=(15, 7))
dendrogram(treeComp, labels=df.Team.to_list())
plt.title("Hierarchical Clustering Dendrogram (Complete Linkage)")
plt.xticks(rotation=35)
plt.xlabel("Teams")
plt.ylabel("Distance")
plt.show()
```

```python
# Hierarchical clustering with complete linkage
treeAve = linkage(d, method='average')

# Plot the dendrogram
plt.figure(figsize=(15, 7))
dendrogram(treeAve, labels=df.Team.to_list())
plt.title("Hierarchical Clustering Dendrogram (Average Linkage)")

plt.xlabel("Teams")
plt.xticks(rotation = 35)
plt.ylabel("Distance")
plt.show()
```

```python
# Hierarchical clustering with complete linkage
treeSingle = linkage(d, method='single')


plt.figure(figsize=(15, 7))
dendrogram(treeSingle, labels=df.Team.to_list())
plt.title("Hierarchical Clustering Dendrogram (Single Linkage)")

plt.xlabel("Teams")
plt.xticks(rotation = 35)
plt.ylabel("Distance")
plt.show()
```

```python
X=ligue1_scaled.to_numpy()
```

```python
n_clusters=2
```

```python
print('============ Influence of the used distance ==============')
for metric in ("l2", "l1", "cosine"): #euclidean
    # clustering
    model = cluster.AgglomerativeClustering(n_clusters=n_clusters, linkage="complete", affinity=metric).fit(X)
    # plot dendogram
    plot_dendrogram(model)
    # assigned cluster to the samples (kind of labelling)
    assigned_cluster = model.labels_
    # trace des clusters obtenus
    plot_clusters(X, assigned_cluster, title='Clustering - linkage complete - Metric {}'.format(metric), symbolsize=80)
```

```python
print('============ Influence of the linkage metric ==============')
n_clusters = 2;
for linkage in ("ward", "complete", "single"):
    # clustering
    model = cluster.AgglomerativeClustering(n_clusters=n_clusters, linkage=linkage, affinity="euclidean").fit(X)
    # plot dendogram
    plot_dendrogram(model)
    # assigned cluster to the samples (kind of labelling)
    assigned_cluster = model.labels_
    # trace des clusters obtenus
    plot_clusters(X, assigned_cluster, title='Clustering - linkage = {} - Metric L2'.format(linkage), symbolsize=80)
```

### K-Means

```python
vect_n_clusters = np.arange(2,12)

from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

silhouette_index = np.empty(vect_n_clusters.shape)
inertia_index = np.empty(vect_n_clusters.shape)

for idx, n_clusters in enumerate(vect_n_clusters):
    #model = cluster.AgglomerativeClustering(n_clusters=n_clusters, linkage="complete", affinity="l2")
    model = cluster.KMeans(n_clusters = n_clusters, init='random')
    
    # assigned cluster to the samples (kind of labelling)
    assigned_cluster = model.fit_predict(X)

    # compute the silhouettte score (higher is the score, better it is)
    sil_avg = silhouette_score(X, assigned_cluster)
    silhouette_index[idx] =  sil_avg
    # comment the line below if using hierarchical clustering
    #inertia_index[idx] = model.inertia_
    
    print("For n_clusters = ,", n_clusters, "the average silhouette_score is :", sil_avg)
    
    # show the clusters
    plot_clusters(X, assigned_cluster, title="K = {}".format(n_clusters), symbolsize=5)
    
plt.figure(figsize=(8,6)); plt.subplot(121)
plt.plot(vect_n_clusters, silhouette_index, '*--'); plt.title('Silhouette score'); plt.xlabel('K')
# comment the line below if using hierarchical clustering
#plt.subplot(122); plt.plot(vect_n_clusters, inertia_index, '*--'); plt.title('Inertia criterion'); plt.xlabel('K')
```
